```python
from proj_tools import *
import wuml
import numpy as np
import torch
import torch.nn as nn
from scipy import stats
```

```python
data = wuml.wData(xpath='./data/D3_Imputed_Balanced_regression.csv', 
					label_type='discrete', label_column_name='preterm_best', first_row_is_label=True)
```

```python
X0 = data.get_all_samples_from_a_class(0)
X1 = data.get_all_samples_from_a_class(1)
```

```python
key_factors = ['MEP' ,'BP3' ,'TCS' ,'PPB' ,'MIBP' ,'MPB' ,'MECPP' ,'MBP' ,'stress', 'BMI_prepreg', 'DCP25', 'TCB', 'age_m', 'MHIBP', 'MEOHP', 'MCOP', 'MEHHP', 'MBZP']
for factor in key_factors:
	print('\nT test for %s'%factor)
	MEP0 = np.squeeze(X0.get_columns([factor]).X)
	MEP1 = np.squeeze(X1.get_columns([factor]).X)
	Tstats = stats.ttest_ind(MEP0, MEP1)
	
	print('\tNormal Birth mean %s = %.4f ± %.3f'%	(factor, np.mean(MEP0), np.std(MEP0) ))
	print('\tNormal Birth median %s = %.4f'%	(factor, np.mean(MEP0) ))
	print('\tAbnormal Birth mean %s = %.4f ± %.3f'%	(factor, np.mean(MEP1), np.std(MEP1) ))
	print('\tAbnormal Birth median %s = %.4f'%	(factor, np.mean(MEP1) ))
	if Tstats[1] < 0.05:
		print('\tp value: %.4f <------------- significant'%Tstats[1])
	else:
		print('\tp value: %.4f'%Tstats[1])
```
