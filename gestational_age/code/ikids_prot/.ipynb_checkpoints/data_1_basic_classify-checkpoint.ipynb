```python
import wuml
from wuml.IO import *
import numpy as np
import torch
import wplotlib
import torch.nn as nn
 
```

```python
data = wuml.wData(xpath='./data/data_1_imputed.csv', batch_size=32, preprocess_data='center and scale',
					label_type='continuous', label_column_name='gestationAge',
					first_row_is_label=True)
```

```python
wuml.jupyter_print('\n\nRun all regressors sorted by least test error')
result = wuml.run_every_regressor(data, alpha=0.1, gamma=0.05, l1_ratio=0.05)
wuml.jupyter_print(result['Train/Test Summary'])
```

```python
cnames = data.get_column_names_as_a_list()
result['RandomForest'].plot_feature_importance('Feature Importance for Random Forest', cnames, xticker_rotate=90, ticker_fontsize=6)
result['RandomForest'].output_sorted_feature_importance_table(data.columns)
```

```python
result['Lasso'].plot_feature_importance('Feature Importance for Lasso', cnames, xticker_rotate=90, ticker_fontsize=6)
result['Lasso'].output_sorted_feature_importance_table(data.columns)
```

```python
YvY = result['RandomForest'].show_true_v_predicted()
wuml.jupyter_print(YvY.df, display_all_rows=True)
```
