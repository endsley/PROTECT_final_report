```python
data = wuml.wData('../data/Chem_decimated_imputed.csv', label_column_name='finalga_best', 
					label_type='continuous', row_id_with_label=0)
```

Create network and train

```python
bNet = wuml.basicNetwork('mse', data, networkStructure=[(200,'relu'),(200,'relu'),(200,'relu'),(1,'none')], max_epoch=6000, learning_rate=0.001)
bNet.train(print_status=True)
Ŷ = bNet(data, output_type='ndarray')		#Takes Numpy array or Tensor as input and outputs a Tensor
```

Check out our predictions

```python
SR = wuml.summarize_regression_result(data.Y, Ŷ)
wuml.jupyter_print(SR.true_vs_predict(sort_based_on_label=True), display_all=True)
```
