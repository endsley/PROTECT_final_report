```python
import os
import sys
if os.path.exists('/home/chieh/code/wuML'):
	sys.path.insert(0,'/home/chieh/code/wuML')
import wuml 
```

```python
data = wuml.wData(xpath='../../data/data.comp3.csv', batch_size=20, 
					label_type='continuous', label_column_name='finalga_best', 
					row_id_with_label=0, columns_to_ignore=['id'])
```

```python
wuml.jupyter_print('\n\nRun all regressors sorted by least test error')
result = wuml.run_every_regressor(data, alpha=0.1, gamma=0.05, l1_ratio=0.05)
```

```python
result['Train/Test Summary']
```

```python
cnames = data.get_column_names_as_a_list()
result['RandomForest'].plot_feature_importance('Feature Importance for Random Forest', cnames, xticker_rotate=90, ticker_fontsize=6)
result['RandomForest'].output_sorted_feature_importance_table(data.columns)
```

```python
result['linear'].plot_feature_importance('Feature Importance for Linear Regression', cnames,xticker_rotate=90, ticker_fontsize=6)
result['linear'].output_sorted_feature_importance_table(data.columns)
```

```python
result['Elastic net'].plot_feature_importance('Feature Importance for Elastic Net', cnames,xticker_rotate=90, ticker_fontsize=6)
result['Elastic net'].output_sorted_feature_importance_table(data.columns)
```

```python
result['Lasso'].plot_feature_importance('Feature Importance for Lasso', cnames,xticker_rotate=90, ticker_fontsize=6)
result['Lasso'].output_sorted_feature_importance_table(data.columns)
```

```python
result['Ridge'].plot_feature_importance('Feature Importance for Ridge Regression', cnames,xticker_rotate=90, ticker_fontsize=6)
result['Ridge'].output_sorted_feature_importance_table(data.columns)
```

```python
result['AdaBoost'].plot_feature_importance('Feature Importance for Ridge Regression', cnames,xticker_rotate=90, ticker_fontsize=6)
result['AdaBoost'].output_sorted_feature_importance_table(data.columns)
```

```python
YvY = result['RandomForest'].show_true_v_predicted()
wuml.jupyter_print(YvY.df, display_all_rows=True)
```
